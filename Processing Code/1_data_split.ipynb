```python
import pandas as pd
from config import path, data_raw, data_target, data_features
```

```python
df = pd.read_csv(path + data_raw).astype(int, errors='ignore')
df.info()
```

```python
df.head()
```

```python
def lead_zeros(x):
    return str('0'*(4-len(str(x))) + str(x))


def frames(x):
    url = x.split('__video-frame')[0]
    time = int(x.split('__video-frame')[1].split('.')[0])
    return str(url + ':' + lead_zeros(time))


def video(x):
    return str(x.split('__video-frame')[0])


frame = df['picName'].apply(frames)
url = df['picName'].apply(video)
```

```python
df.insert(loc=0, column='frame', value=frame)
df.insert(loc=1, column='url', value=url)
df.insert(loc=2, column='is_food', value=int(0))
df = df.drop(columns='picName')
df = df.sort_values(by=['frame'])
df.info()
```

```python
df.head()
```

```python
df[['frame', 'url', 'is_food']].to_csv(path + data_target, index=False)
df.drop(columns=['frame', 'is_food']).to_csv(
    path + data_features, index=False)
```
