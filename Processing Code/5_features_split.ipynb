```python
import pandas as pd
import os
from config import path, features, features_path, file
```

```python
df = pd.read_csv(path + features).astype(float, errors='ignore')
url_list = df['url'].unique()
df.head()
```

```python
if not os.path.exists(features_path):
    os.makedirs(features_path)

i = 0
for url in url_list:
    i += 1
    raw = df.loc[df['url'] == url]
    raw.drop(columns=['url']).to_csv(features_path + file(url), index=False, header=False)
    print(i, '/', len(url_list))
```
