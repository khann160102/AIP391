```python
import pandas as pd
import numpy as np
from config import path, data_features, statistics
```

```python
df = pd.read_csv(path + data_features).drop(columns='url').astype(int, errors='ignore')
df = df.sort_index(axis=1)
df.head()
```

```python
count = np.empty(shape=(len(df.columns)), dtype=int)
sum = np.empty(shape=(len(df.columns)))
median = np.empty(shape=(len(df.columns)), dtype=int)
i = 0
for color in df.columns:
    count[i] = np.count_nonzero(df[color])
    sum[i] = np.sum(df[color])
    median[i] = np.median(df[color])
    i += 1
```

```python
stats = df.describe().transpose()
stats['count'] = count
stats.insert(loc=1, column='sum', value=sum)
stats.insert(loc=2, column='median', value=median)
stats
```

```python
stats.to_excel(path + statistics)
```
